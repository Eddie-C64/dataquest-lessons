# A look at the data

Before we get started with plotting, let's take a look at our data.  Each row in the data contains information on how a player performed in the 2013-2014 NBA season.

Here are some selected columns:

```python
import pandas
with open("nba_2013.csv", 'r') as csvfile:
    nba = pandas.read_csv(csvfile)

# The names of the columns in the data.
print(nba.columns.values)
```

# Plotting with seaborn

Let's make some plots with seaborn.  The first type of plot we make will be an `lmplot`, which in seaborn gives you a scatterplot with a regression line on top of it.  This makes it easy to explore the relationship between two variables.

## Instructions

The `x_partial` argument to the `lmplot` function can be used to remove the effect of a confounding variable.  `sns.lmplot("pf", "blk", df, x_partial="drb")` will plot the `pf` column against the `blk` column with any confounding effect of the `drb` column removed.

Use the `x_partial` argument to plot `pts` against `ast` with minutes played (`mp`) removed.

## Hint

Add the named keyword argument `x_partial` to a call to `sns.lmplot`, and set `x_partial` to `mp`.

```python
## Initial

import random
random.seed(1)

## Display

%matplotlib inline
import seaborn as sns

# We make an lmplot by specifying the names of columns we want to plot.
# We also specify the pandas dataframe the columns are found in.
sns.lmplot("pts", "ast", nba)
sns.plt.show()

## Answer

import numpy
random.seed(1)
numpy.random.seed(1)
sns.lmplot("pts", "ast", nba, x_partial="mp")
sns.plt.show()

## Check vars

["dq_plot2"]
```

# Plotting the residuals

We can also easy plot the residuals that result from fitting a regression line to two columns of data.  This involves the `residplot` function.  We pass it two pandas series or numpy vectors, and it makes a plot of the residuals.

Because `residplot` is a lower level plotting function, we need to add a call to `sns.plt.show` after calling it to show the plot properly.

## Instructions

We added a column to the data called `ts.`.  This is the true shooting percentage, a measure of how well a player shot the ball while accounting for the effect of free throws and field goals.

Make a `residplot` of the `ts.` column against the `mp` column (`ts.` column on the x-axis).

## Hint

Call the `sns.residplot` function, but change the columns you pass in.

```python
## Initial

nba["ts."] = nba["pts"] / (2 * (nba["fga"] + .44 * nba["fta"]))

## Display

# The residuals get larger as the minutes increase.
# It's harder to predict the point total of someone who plays a lot of minutes.
sns.residplot(nba["mp"], nba["pts"])
sns.plt.show()

## Answer

sns.residplot(nba["ts."], nba["mp"])

## Check vars

["dq_plot2"]
```
